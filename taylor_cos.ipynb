Можно посчитать косинус угла через ряд Тейлора, используя то, что мы можем посчитать значение производной любого порядка в точке 0.
$$cosx=1+0-\frac{x^2}{2!}+0+\frac{x^4}{4!}...=1-\frac{x^2}{2!}+\frac{x^4}{4!}-\frac{x^6}{6!}...$$

```python
import math
import matplotlib.pyplot as mpp
import numpy

n=55

def t_cos(x):
    sum=1
    num=1
    den=1
    pow=2
    for i in range(0,n):
        num*=(pow-1)*pow
        den*=(pow-1)*pow
        if pow%4==0:
            sum=(num+x**pow)/den
        else:
            sum=(num-x**pow)/den
        pow+=2
        num=sum*den
    return sum

print(math.cos(0.7))
print(t_cos(0.7))
```

Теперь можно построить визуальное сравнение библиотечного и полученного косинусов.

```python
vec = numpy.vectorize(t_cos)
arguments = numpy.arange(-20,20,0.5)
mpp.plot(arguments, numpy.cos(arguments))
mpp.plot(arguments, vec(arguments))
mpp.show()
```

Из график видно, что при сравнительно малых углах библиотечный и полученный косинусы совпадают. Интересно теперь провести такое же сравнение, но уже при больших углах.

```python
vec = numpy.vectorize(t_cos)
arguments = numpy.arange(-50,50,0.1)
mpp.plot(arguments, numpy.cos(arguments))
mpp.plot(arguments, vec(arguments))
mpp.show()
```

При больших углах уже заметны расхождения между значениями функций.
